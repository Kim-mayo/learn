```python
import random

# 이분 탐색  ( 정렬된 상태에서 사용 가능)
def binary_search_recursive(data, search):
    if len(data) == 0:
        return False
    if len(data) == 1 and search == data[0]:
        return True
    elif len(data) == 1 and search != data[0]:
        return False
    
    mid = len(data)//2
    if data[mid] == search:
        return True
    elif data[mid] > search:
        return binary_search(data[:mid],search)
    else:
        return binary_search(data[mid+1:],search)

def binary_search_iterated(data, search):
    if len(data) ==0:
        return False
    low, high = 0, len(data)-1
    while high >= low:
        mid = (low+high)//2
        if data[mid] == search:
            return True
        elif data[mid] < search:
            low = mid+1
        else:
            high = mid-1
    # 못찾은 경우
    return False
# search 값 이상인 값의 index 반환
def lower_bound(data, search):
    if len(data) <= 0 :
        return None
    
    low , high = 0 , len(data)
    while high > low:
        mid = (low+high)//2
        if data[mid] < search:
            low = mid+1
        else:
            high = mid
    
    return high
# search 값 초과한 값의 index 반환
def upper_bound(data, search):
    if len(data) <= 0:
        return None
    
    low , high = 0, len(data)
    while high > low:
        mid = (low+high)//2
        if data[mid] <= search:
            low = mid+1
        else:
            high = mid
    return high

if __name__ == "__main__":
    data = random.sample(range(1000), 100)
    data2 = random.sample(range(2000), 100)
    data.extend(data2)
    data = sorted(data)
    print(data)
    for i in range(5):
        search = random.randint(0,1000)
        print("Search 값 = {0}".format(search))
        print(binary_search_recursive(data, search) , end =" ")
        print(binary_search_iterated(data, search), end = " ")
        print(lower_bound(data, search) , end = " ")
        print(upper_bound(data, search))
```
