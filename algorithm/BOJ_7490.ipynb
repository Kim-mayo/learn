```python
def zero_make(operator, n):
    if(len(operator) == n-1):
        Sum = [1]
        num = 2
        # 공백 문자일 경우 숫자 합쳐줘야한다.
        for op in operator:
            if op == ' ':
                Sum[-1] = Sum[-1]*10 + num
            else:
                Sum.append(op)
                Sum.append(num)
            num+=1
        # + 인지 - 인지
        result = Sum[0]
        for _ in range(1,len(Sum)):
            if Sum[_] == '+' or Sum[_]=='-':
                continue
            else:
                if Sum[_-1] == '+':
                    result += Sum[_]
                else:
                    result -= Sum[_]
        if result == 0 :
            print('1' , end="")
            num = 2
            for op in operator:
                print(op, end="")
                print(num, end="")
                num+=1
            print()
        return        
            
    operator.append(' ')
    zero_make(operator,n)
    operator[-1] = '+'
    zero_make(operator, n)
    operator[-1] = '-'
    zero_make(operator,n)
    operator.pop()
    
    return

T =int(input())
for _ in range(T):
    n = int(input())
    operator = []
    zero_make(operator, n)
    print()
```
