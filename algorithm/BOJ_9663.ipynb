```python
count = 0
def NQueen(curr_row, max_row , curr_pick):
    if curr_row == max_row:
        global count
        count +=1
        return
    
    for curr_col in range(max_row):
        available = True
        for index in range(len(curr_pick)):
            if curr_pick[index] == curr_col or abs(curr_row-index) == abs(curr_col-curr_pick[index]):
                available = False
                break
        if available == True:
            curr_pick.append(curr_col)
            NQueen(curr_row +1 , max_row, curr_pick)
            curr_pick.pop()
            
    return
    
N = int(input())
NQueen(0,N,[])
print(count)
```
